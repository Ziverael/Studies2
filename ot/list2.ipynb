```python
import numpy as np
```

```python
def netwon_method(df, ddf, x0: float, eps: float, M: int) -> float:
    x_ = x0
    for step in range(1, M + 1):
        x0 = x_
        print(f"{step=},x*{round(x_, 3)},x0={round(x0, 3)}")
        change = df(x_) / ddf(x_)
        x_ = x_ - change
        print(f"{change=}")
        print("-" * 2)
        if abs(x0 - x_) < eps:
            break
    return round(x_, 3)
```

```python
netwon_method(
    lambda x: -3*x**2+8*x+1,
    lambda x: -6*x+8,
    -1,
    0.001,
    3
)
```

```python
netwon_method(
    lambda x: -3*x**2+8*x+1,
    lambda x: -6*x+8,
    2,
    0.001,
    3
)
```

```python
def secant_method(df, x0: float, x1: float, eps: float, M: int) -> float:
    for step in range(1, M + 1):
        df0, df1  = df(x0), df(x1)
        x_ = (x1 - x0) * df1 / (df1 - df0)
        x0 = x1
        x1 = x_
        print(f"{step=},x0={round(x0, 3)},x1={round(x1, 3)}")
        print("-" * 2)
        if abs(x0 - x1) < eps:
            break
    return round(x_, 3)
```

```python
secant_method(
    lambda x: -3*x**2+8*x+1,
    -1,
    2,
    0.001,
    12
)
```
